```python
cookie={} # insert request cookies within{}
def getAmazonSearch(search_query):
    url="https://www.amazon.in/s?k="+search_query
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
def Searchasin(asin): #A function to get the contents of individual product pages using 'data-asin' number (unique identification number)
    url="https://www.amazon.in/dp/"+asin
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
def Searchreviews(review_link):#A function to pass on the link of 'see all reviews' and extract the conten
    url="https://www.amazon.in"+review_link
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
#page product reviews extraction
product_names=[]
response=getAmazonSearch('mobile')
soup=BeautifulSoup(response.content)
for page in range(1,6):

    for i in soup.findAll("span",{'class':'a-size-medium a-color-base a-text-normal'}):# the tag which is common for all the names of products
        product_names.append(i.text) #adding the product names to the list
```

```python

data_asin=[]
response=getAmazonSearch('mobile')
soup=BeautifulSoup(response.content)
for page in range(1,6):
    for i in soup.findAll("div",{'class':"sg-col-20-of-24 s-result-item s-asin sg-col-0-of-12 sg-col-28-of-32 sg-col-16-of-20 sg-col sg-col-32-of-36 sg-col-12-of-16 sg-col-24-of-28"}):
        data_asin.append(i['data-asin'])
```

```python
data_asin
```

```python
len(data_asin)
```

```python
link=[]
for i in range(len(data_asin)):
    response=Searchasin(data_asin[i])
    soup=BeautifulSoup(response.content)
    for i in soup.findAll("a",{'data-hook':"see-all-reviews-link-foot"}):
        link.append(i['href'])
```

```python
len(link)
```

```python

reviews=[]
for j in range(len(link)):
    for k in range(10):
        response=Searchreviews(link[j]+'&pageNumber='+str(k))
        soup=BeautifulSoup(response.content)
        for i in soup.findAll("span",{'data-hook':"review-body"}):
            reviews.append(i.text)
```

```python
rev={'reviews':reviews}#converting the reviews list into a dictionary
```

```python
review_data=pd.DataFrame.from_dict(rev)#converting this dictionary into a dataframe
```
